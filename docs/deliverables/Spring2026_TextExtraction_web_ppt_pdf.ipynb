### LP1: Information from ScrapethisSite

#### Extract Hockey Team Data from ScrapethisSite
'https://www.scrapethissite.com/pages/forms/'


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.scrapethissite.com/pages/forms/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')

table = soup.find('table', class_='table')

headers = []
data = []

if table:
    # Extract headers from the first row of <th> elements
    header_row = table.find('tr') # Assuming the first tr contains headers
    if header_row:
        for th in header_row.find_all('th'):
            headers.append(th.text.strip())

    # Extract data rows, skipping the header row
    for row in table.find_all('tr')[1:]:
        row_data = []
        for td in row.find_all('td'):
            row_data.append(td.text.strip())
        data.append(row_data)

df = pd.DataFrame(data, columns=headers)
print(df.head())
```

### LP4: Extract Table from PPT

```python
!pip install python-pptx
```

```python
from pptx import Presentation
import pandas as pd

prs = Presentation('/content/Your big idea.pptx')
all_tables_data = []

# Iterate through each slide in the presentation
for slide_idx, slide in enumerate(prs.slides):
    # Iterate through every shape in slide.shapes
    for shape_idx, shape in enumerate(slide.shapes):
        # Check if the current shape has a table
        if shape.has_table:
            print(f"Found table on Slide {slide_idx + 1}, Shape {shape_idx + 1}")
            table = shape.table

            current_table_data = []
            if len(table.rows) > 0:
                header_row = [cell.text for cell in table.rows[0].cells]
                current_table_data.append(header_row)

                for row_idx, row in enumerate(table.rows):
                    if row_idx == 0:  # Skip the header row
                        continue
                    row_texts = [cell.text for cell in row.cells]
                    current_table_data.append(row_texts)

                all_tables_data.append(current_table_data)

# check if empty
if not all_tables_data:
    print("No tables were found in the presentation.")
    print("To access tables if they were present, the code iterates through `prs.slides` and then `slide.shapes`. If `shape.has_table` is True, `shape.table` provides access to the table object, from which rows and cells can be iterated to extract text.")
else:
    for i, table_data in enumerate(all_tables_data):
        if table_data and len(table_data) > 1:
            df = pd.DataFrame(table_data[1:], columns=table_data[0])
            print(f"\nExtracted Table {i + 1} (first 5 rows):")
            print(df.head())
        else:
            print(f"\nTable {i + 1} was found but contained no data beyond headers, or was empty.")
```

###HP 1.1: IBM AI Agents

`https://www.ibm.com/think/topics/ai-agents#7281535`

```python
import requests

# URL
ibm_url = 'https://www.ibm.com/think/topics/ai-agents#7281535'

# User-Agent header to mimic a web browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

ibm_response = requests.get(ibm_url, headers=headers)

print(f"Status Code: {ibm_response.status_code}")
print(f"Response content length: {len(ibm_response.text)} characters")
```

```python
from bs4 import BeautifulSoup

ibm_soup = BeautifulSoup(ibm_response.text, 'lxml')

print("BeautifulSoup object 'ibm_soup' created successfully.")
```

```python
print(ibm_soup.prettify())
```

```python
import re

# Extract the main title
main_title_element = ibm_soup.find('h1', class_='cmp-leadspace__heading')
main_title = main_title_element.get_text(strip=True) if main_title_element else "Main title not found"
print(f"Main Title: {main_title}\n")

# Extract content section by section
content_sections = ibm_soup.find_all('div', class_=['rich-text text', 'standalone-title enhanced-title text'])

article_content = []
current_section_title = ""

for section in content_sections:
    # headings
    heading_element = section.find(['h2', 'h3'], class_=re.compile(r'expressive-heading'))
    if heading_element:
        current_section_title = heading_element.get_text(strip=True)
        article_content.append(f"\n## {current_section_title}")

    # Extract paragraphs within the current section
    rich_text_div = section.find('div', class_='cms-richtext')
    if rich_text_div:
        paragraphs = rich_text_div.find_all('p')
        for p in paragraphs:
            # Clean up reference numbers like [1] or [1-2]
            cleaned_text = re.sub(r'\s*\[\d+(?:-\d+)?\]|\s*\d+\s*$', '', p.get_text(separator=' ', strip=True))
            if cleaned_text:
                article_content.append(cleaned_text)

        # Extract list items if present
        list_items = rich_text_div.find_all('li')
        for li in list_items:
            cleaned_text = re.sub(r'\s*\[\d+(?:-\d+)?\]|\s*\d+\s*$', '', li.get_text(separator=' ', strip=True))
            if cleaned_text:
                article_content.append(f"- {cleaned_text}")

# Print the extracted content
for line in article_content:
    print(line)
```

### HP 1.2: Fetch Indietraveller Page Content
 `https://www.indietraveller.co/top-cheap-places-to-travel/`

```python
import requests

# URL
indietraveller_url = 'https://www.indietraveller.co/top-cheap-places-to-travel/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

indietraveller_response = requests.get(indietraveller_url, headers=headers)

print(f"Successfully fetched content from {indietraveller_url} with status code {indietraveller_response.status_code}.")
print(f"Response content length: {len(indietraveller_response.text)} characters.")
```

```python
from bs4 import BeautifulSoup

indietraveller_soup = BeautifulSoup(indietraveller_response.text, 'lxml')

print("BeautifulSoup object 'indietraveller_soup' created successfully.")
```

```python
print(indietraveller_soup.prettify())
```

```python
import re

main_title_element = indietraveller_soup.find('h1', class_='post-title')
main_article_title = main_title_element.get_text(strip=True) if main_title_element else "Main article title not found."

# Initialize a list to store destination data
destinations_data = []

current_destination = None
current_description_paragraphs = []
current_backpacker_budget = None
current_mid_range_budget = None
current_when_to_go = None

entry_content_div = indietraveller_soup.find('div', class_='entry-content')

if entry_content_div:
    # Check all direct children of the 'entry-content' div
    for element in entry_content_div.children:
        # Check if destination heading
        if element.name == 'h2' and 'wp-block-heading' in element.get('class', []):

            if current_destination:
                destinations_data.append({
                    'name': current_destination,
                    'description': ' '.join(current_description_paragraphs).strip(),
                    'backpacker_budget': current_backpacker_budget,
                    'mid_range_budget': current_mid_range_budget,
                    'when_to_go': current_when_to_go
                })

            current_destination = element.get_text(strip=True)
            current_description_paragraphs = []
            current_backpacker_budget = None
            current_mid_range_budget = None
            current_when_to_go = None

        elif current_destination and element.name == 'p':
            p_text = element.get_text(strip=True)
            if p_text:
                backpacker_match = re.search(r'Backpacker Budget:\s*([^\n<]+?)(?:\s*Mid-range Budget:|$)', p_text, re.IGNORECASE)
                if backpacker_match:
                    current_backpacker_budget = backpacker_match.group(1).strip()

                mid_range_match = re.search(r'Mid-range Budget:\s*([^\n<]+?)(?:\s*When to go:|$)', p_text, re.IGNORECASE)
                if mid_range_match:
                    current_mid_range_budget = mid_range_match.group(1).strip()

                when_to_go_match = re.search(r'When to go:\s*([^\n<]+?)(?:\s*READ MORE|$)', p_text, re.IGNORECASE)
                when_match = re.search(r'When:\s*([^\n<]+?)(?:\s*READ MORE|$)', p_text, re.IGNORECASE)

                if when_to_go_match:
                    current_when_to_go = when_to_go_match.group(1).strip()
                elif when_match:
                    current_when_to_go = when_match.group(1).strip()

                if not (("Backpacker Budget:" in p_text or "Mid-range Budget:" in p_text or "When to go:" in p_text or "When:" in p_text) and
                        (backpacker_match or mid_range_match or when_to_go_match or when_match)):
                    current_description_paragraphs.append(p_text)

        elif current_destination and (element.name == 'hr' or (element.name == 'h2' and 'wp-block-heading' not in element.get('class', []))):
            if current_destination:
                destinations_data.append({
                    'name': current_destination,
                    'description': ' '.join(current_description_paragraphs).strip(),
                    'backpacker_budget': current_backpacker_budget,
                    'mid_range_budget': current_mid_range_budget,
                    'when_to_go': current_when_to_go
                })
                current_destination = None
                current_description_paragraphs = [] # Reset

    if current_destination:
        destinations_data.append({
            'name': current_destination,
            'description': ' '.join(current_description_paragraphs).strip(),
            'backpacker_budget': current_backpacker_budget,
            'mid_range_budget': current_mid_range_budget,
            'when_to_go': current_when_to_go
        })

# Print the extracted data
print(f"Main Article Title: {main_article_title}")
print("\n--- Sample of Extracted Destinations ---")
for i, dest in enumerate(destinations_data[:5]):
    print(f"\nDestination {i+1}:")
    print(f"  Name: {dest['name']}")
    print(f"  Description: {dest['description'][:150]}...")
    print(f"  Backpacker Budget: {dest['backpacker_budget']}")
    print(f"  Mid-Range Budget: {dest['mid_range_budget']}")
    print(f"  When to Go: {dest['when_to_go']}")

print(f"\nTotal destinations extracted: {len(destinations_data)}")
```

```python
!pip install scrapy
import scrapy

class WikiSpider(scrapy.Spider):
    name = 'wiki_sf_history'
    start_urls = [
        'https://lahistorygirl.com/my-ultimate-guide-to-san-francisco/'
    ]

    custom_settings = {
        'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    def parse(self, response):
        # Extract the main title of the article
        main_title = response.css('h1.entry-title::text').get()
        if main_title:
            yield {
                'type': 'main_title',
                'content': main_title.strip()
            }

        # Extract content from the main article body
        # Looking for headings and paragraphs within the 'entry-content' div
        for selector in response.css('div.entry-content > *'):
            if selector.css('h2::text').get():
                yield {
                    'type': 'heading_2',
                    'content': selector.css('h2::text').get().strip()
                }
            elif selector.css('h3::text').get():
                yield {
                    'type': 'heading_3',
                    'content': selector.css('h3::text').get().strip()
                }
            elif selector.css('p::text').getall():
                # Join all text nodes within a paragraph
                paragraph_text = " ".join(selector.css('p::text').getall()).strip()
                if paragraph_text:
                    yield {
                        'type': 'paragraph',
                        'content': paragraph_text
                    }
            elif selector.css('ul li::text').getall():
                list_items = [li.strip() for li in selector.css('ul li::text').getall() if li.strip()]
                if list_items:
                    yield {
                        'type': 'list_items',
                        'content': list_items
                    }

# Save the spider code to a file
spider_code = """
import scrapy

class WikiSpider(scrapy.Spider):
    name = 'wiki_sf_history'
    start_urls = [
        'https://lahistorygirl.com/my-ultimate-guide-to-san-francisco/'
    ]

    custom_settings = {
        'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    def parse(self, response):
        # Extract the main title of the article
        main_title = response.css('h1.entry-title::text').get()
        if main_title:
            yield {
                'type': 'main_title',
                'content': main_title.strip()
            }

        # Extract content from the main article body
        for selector in response.css('div.entry-content > *'):
            if selector.css('h2::text').get():
                yield {
                    'type': 'heading_2',
                    'content': selector.css('h2::text').get().strip()
                }
            elif selector.css('h3::text').get():
                yield {
                    'type': 'heading_3',
                    'content': selector.css('h3::text').get().strip()
                }
            elif selector.css('p::text').getall():
                # Join all text nodes within a paragraph
                paragraph_text = " ".join(selector.css('p::text').getall()).strip()
                if paragraph_text:
                    yield {
                        'type': 'paragraph',
                        'content': paragraph_text
                    }
            elif selector.css('ul li::text').getall():
                list_items = [li.strip() for li in selector.css('ul li::text').getall() if li.strip()]
                if list_items:
                    yield {
                        'type': 'list_items',
                        'content': list_items
                    }
"""

with open('quotes_spider.py', 'w') as f:
    f.write(spider_code)

print("Scrapy spider code saved to quotes_spider.py")
```

```python
import requests

# URL
url = 'https://en.wikipedia.org/wiki/World_Tourism_rankings'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

wiki_response = requests.get(url, headers=headers)

print(f"Successfully fetched content from {url} with status code {wiki_response.status_code}.")
```

```python
from bs4 import BeautifulSoup

wiki_soup = BeautifulSoup(wiki_response.text, 'lxml')

print("BeautifulSoup object 'wiki_soup' created successfully.")
```

```python
print(wiki_soup.prettify())
```

```python
import pandas as pd
import re
from bs4 import BeautifulSoup

# Find all tables with 'wikitable' and 'sortable' classes
tables = wiki_soup.find_all('table', class_=['wikitable', 'sortable'])

all_extracted_tables = {}

for i, table in enumerate(tables):
    caption_tag = table.find('caption')
    raw_caption = caption_tag.get_text(strip=True) if caption_tag else f"Table {i+1}"
    # remove reference numbers for consistent dictionary keys
    caption_key = re.sub(r'\s*\[[a-zA-Z0-9]+\]', '', raw_caption).strip()

    current_headers = []
    if "International tourist arrivals (in millions)" in caption_key:
        current_headers = ["Destination", "2024_Arrivals", "Change_2023_2024_percent", "2023_Arrivals", "2022_Arrivals"]
    elif "International tourism receipts (US$billions)" in caption_key:
        current_headers = ["Region", "2024_Receipts_USD_billions", "2023_Receipts_USD_billions", "Change_local_currency_percent"]
    elif "International tourism expenditure (US$ billion)" in caption_key:
        current_headers = ["Country", "2024_Expenditure_USD_billions", "2023_Expenditure_USD_billions", "2022_Expenditure_USD_billions"]
    else:
        header_row_ths = table.find('tr').find_all('th')
        current_headers = [re.sub(r'\s*\[[a-zA-Z0-9]+\]', '', th.get_text(strip=True)).strip() for th in header_row_ths if th.get_text(strip=True)]
        current_headers = [h for h in current_headers if h] # Remove empty strings

    if not current_headers:
        print(f"Skipping table '{raw_caption}' due to missing or unidentifiable headers.")
        continue

    rows_data_for_df = []
    all_tr_elements = table.find_all('tr')
    start_row_index = 0
    if len(all_tr_elements) > 0 and all(cell.name == 'th' for cell in all_tr_elements[0].find_all(['td', 'th'])):
        start_row_index = 1
    if ("International tourist arrivals (in millions)" in caption_key or
        "International tourism receipts (US$billions)" in caption_key or
        "International tourism expenditure (US$ billion)" in caption_key) and start_row_index == 0:
        start_row_index = 1


    for tr_element in all_tr_elements[start_row_index:]:
        cols = []
        for cell in tr_element.find_all(['td', 'th']):
            link_text = cell.find('a')
            if link_text:
                cols.append(link_text.get_text(strip=True))
            else:
                text = cell.get_text(separator=' ', strip=True)
                cleaned_text = re.sub(r'\s*\[[a-zA-Z0-9]+\]', '', text).strip()
                cols.append(cleaned_text)

        if any(c is not None and c != '' for c in cols):
            if len(cols) > len(current_headers):
                cols = cols[:len(current_headers)]
            elif len(cols) < len(current_headers):
                cols = cols + [None] * (len(current_headers) - len(cols))

            rows_data_for_df.append(cols)

    if rows_data_for_df and current_headers and len(rows_data_for_df[0]) == len(current_headers):
        all_extracted_tables[caption_key] = pd.DataFrame(rows_data_for_df, columns=current_headers)
    else:
        print(f"No valid data rows or header length mismatch for table '{raw_caption}'. Expected {len(current_headers)} columns, but data rows have {len(rows_data_for_df[0]) if rows_data_for_df else 0}.")

# Print the head of each extracted DataFrame
for caption_key_iter, df in all_extracted_tables.items():
    print(f"\n--- {caption_key_iter} ---")
    print(df.head())


if "International tourist arrivals (in millions)" in all_extracted_tables:
    df_tourism_arrivals = all_extracted_tables["International tourist arrivals (in millions)"]
    print("\nDataFrame 'df_tourism_arrivals' created from 'International tourist arrivals (in millions)'.")
else:
    df_tourism_arrivals = pd.DataFrame() # Create an empty DataFrame if not found
    print("Could not find 'International tourist arrivals (in millions)' table.")

if "International tourism receipts (US$billions)" in all_extracted_tables:
    df_tourism_receipts = all_extracted_tables["International tourism receipts (US$billions)"]
    print("\nDataFrame 'df_tourism_receipts' created from 'International tourism receipts (US$billions)'.")
else:
    df_tourism_receipts = pd.DataFrame()
    print("Could not find 'International tourism receipts (US$billions)' table.")

if "International tourism expenditure (US$ billion)" in all_extracted_tables:
    df_tourism_expenditure = all_extracted_tables["International tourism expenditure (US$ billion)"]
    print("\nDataFrame 'df_tourism_expenditure' created from 'International tourism expenditure (US$ billion)'.")
else:
    df_tourism_expenditure = pd.DataFrame()
    print("Could not find 'International tourism expenditure (US$ billion)' table.")
```
